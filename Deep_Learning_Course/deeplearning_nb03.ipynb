## Tutorial 5: *Dropout & Batch Normalization*

Use the *'red-wine'* dataset.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

from tensorflow import keras
from keras import layers, callbacks
from keras.callbacks import EarlyStopping

datapath = os.getcwd()+'/sample_data'
```

### Import and split up the data

```python
ion = pd.read_csv(datapath+'/ion.csv', index_col=0)

ion.describe()
```

```python
df = ion.copy()

df['Class'] = df['Class'].map({'good':0, 'bad':1})
```

```python
df_train = df.sample(frac=0.70, random_state=0)

df_valid = df.copy().drop(df_train.index)
```

```python
print(len(ion))
print()
print(len(df_train))
print()
print(len(df_valid))
```

```python
target = 'Class'

max_ = df_train.max(axis=0)
min_ = df_train.min(axis=0)

df_train = (df_train - min_) / (max_ - min_)
df_train.dropna(axis=1, inplace=True)

df_valid = (df_valid - min_) / (max_ - min_)
df_valid.dropna(axis=1, inplace=True)


X_train = df_train.drop(target, axis=1)
y_train = df_train[target]

X_valid = df_valid.drop(target, axis=1)
y_valid = df_valid[target]
```

```python
display(df_train.head(5))
```

### The Model with new layers

```python
model = keras.Sequential([
    layers.Dense(4, activation='relu', input_shape=[33]),
    layers.Dense(4, activation='relu'),
    layers.Dense(1, activation='sigmoid')])

model.compile(optimizer='adam', 
              loss='binary_crossentropy',
              metrics=['binary_accuracy'])

early_stopping = keras.callbacks.EarlyStopping(patience=10,
                                               min_delta=0.001,
                                               restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data = (X_valid, y_valid),
    batch_size=512,
    epochs=1000,
    callbacks=[early_stopping],
    verbose=0)
```

```python
history_df = pd.DataFrame(history.history)

history_df.loc[5:,['loss','val_loss']].plot()
history_df.loc[5:, ['binary_accuracy','val_binary_accuracy']].plot()

print(("Best Validation Loss: {:0.4f}" +\
       "\nBest Validation Accuracy: {:0.4f}")\
      .format(history_df.val_loss.min(),
              history_df.val_binary_accuracy.max()))
```
